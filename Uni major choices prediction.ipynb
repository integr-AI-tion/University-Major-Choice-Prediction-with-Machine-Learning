# Import lib & load data 

```python
import pandas as pd

df = pd.read_csv("ML project survey responses.csv")
```

# Clean target labels (y)

```python
df = df.rename(
    columns={
        'Kelompok jurusan manakah yang PALING mungkin Anda pilih untuk jenjang sarjana (S1)?': 'chosen_major'
    }
)

target = 'chosen_major'
```

```python
y = df[target]
y_clean = y.apply(lambda x: x.split(',')[0].strip())
y_clean = y_clean.replace({
    "I guess it technically counts as seni/desain??? I know the option is LITERALLY right there but I'm just typing this in because I'm bored.": "Seni / Desain"
})
y_clean.value_counts()
```

# Define x and y

```python
X = df.drop(
    columns=[
        'Timestamp',
        'Nama Lengkap',
        'Kelas',
        'E-mail',
        'Nomor Telepon (untuk e-wallet)',
        target
    ]
)
```

```python
y = y_clean

X.head(), y.head()
```

# Encode X

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=True)
X_encoded = encoder.fit_transform(X)

X_encoded.shape
```

# Splitting data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.25, random_state=42)
```

# Train model (logistic regression)

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000, solver="lbfgs",)
model.fit(X_train, y_train)
```

# Evaluate model

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
y_pred = model.predict(X_test)

print("Accuracy : ", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, zero_division=0))
```

# Confusion matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 10))
display = ConfusionMatrixDisplay.from_predictions(y_test, y_pred, xticks_rotation=60)
plt.xticks(fontsize=9, ha="right")
plt.yticks(fontsize=9)

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```
